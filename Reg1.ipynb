```python
#"¬øla democracia en Argentina es hoy: muy estable, algo estable, poco estable o nada estable?"
x = df_diputados["DEM1"]

print(x.head())
```

```python
#"En una escala de 1 a 10, donde 1 significa "no es una cuesti√≥n importante" y el 10 "es una cuesti√≥n muy importante"? valore el conflicto entre los poderes del Estado"
y = df_diputados["PRO108"]

print(y.head())
```

```python
sns.scatterplot(x='DEM1', y='PRO108', data=df_diputados)

plt.title('Grafico de dispersion entre Estabilidad democr√°tica y Conflicto entre poderes')
plt.xlabel('Estabilidad democr√°tica')
plt.ylabel('Conflicto entre poderes')
```

```python
correlacion_pearson = df_diputados["DEM1"].corr(df_diputados["PRO108"])
print("La correlaci√≥n de Pearson entre Estabilidad democr√°tica y Conflicto entre poderes es:", correlacion_pearson)
```

```python
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

r_squared = model.rsquared
print("El coeficiente de determinaci√≥n (R¬≤) es:", r_squared)
```

```python
covarianza = df_diputados[['DEM1', 'PRO108']].cov()

print("La matriz de covarianza es:")
print(covarianza)
```

```python
print(model.summary())
```

```python
plt.scatter(df_diputados["DEM1"], df_diputados["PRO108"], label='Datos reales', color='blue')

intercepto, pendiente = model.params
x_vals = np.linspace(df_diputados["DEM1"].min(), df_diputados["DEM1"].max(), 100)
y_vals = intercepto + pendiente * x_vals

# Dibujar la recta de regresi√≥n en el gr√°fico de dispersi√≥n
plt.plot(x_vals, y_vals, color='red', label=f'Recta de regresi√≥n (pendiente={pendiente:.2f})')

# Personalizar el gr√°fico
plt.xlabel("Estabilidad democr√°tica")
plt.ylabel("Conflicto entre los poderes del Estado")
plt.title('Regresi√≥n lineal')
plt.legend()

plt.show()
```

```python
print(f'Intercepto : {intercepto:.2f}')
print(f'Pendiente : {pendiente:.2f}')
```

```python
#Correlaci√≥n entre "Estabilidad democratica" y "Conflicto entre los poderes del Estado":
#La correlaci√≥n entre ambas variables es bastante baja, siendo de 0.1 aproximadamente. Como se muestra en el gr√°fico de dispersi√≥n,
#quienes contestaron que la democracia en Argentina era muy estable, no necesariamente reflejaron una tendencia espec√≠fica al contestar sobre la importancia del conflicto entre poderes.
```

```python
#Regresion 
#Teniendo en cuenta el resultado de R2 que del 0.01, siendo "Estabilidad democratica" la variable X (independiente) y "Conflicto entre los poderes del Estado" la variable y (dependiente). Se puede afirmar que x proporciona una baja determinacion de y.
```

```python
#Covarianza
#En la matriz hay una relaci√≥n positiva entre ambas variables, lo que quiere decir que al aumentar la variable "Estabilidad democratica" tambi√©n aumenta la variable "Conflicto entre los poderes del Estado" (esto no significa que haya correlaci√≥n o causalidad).
#La matriz de covarianza es:
            #"Estabilidad democratica"    "Conflicto entre los poderes del Estado"
#"Estabilidad democratica"                 0.303670  0.100917
#"Conflicto entre los poderes del Estado"  0.100917  3.349458
#Sin embargo "Conflicto entre los poderes del Estado" muestra mayor variabilidad que "Estabilidad Democratica". Lo que tambi√©n puede verse en el gr√°fico de dispersi√≥n,
#donde el conjunto de datos de "Estabilidad democratica" se encuentra mayormente en los valores 3 y 4, y el conjunto de datos de "Conflicto entre los poderes del Estado" se encuantran dispersos entre todos los valores posibles (3 a 10).
```

```python
#Significatividad
#No hay suficiente evidencia para afirmar que la variable "Estabilidad democratica" est√© relacionado de forma significativa con la variable "Conflicto entre los poderes del Estado", siendo p-valor mayor a 0.05 (0.298). 
# Este resultado va en consonancia con las conclusiones anteriores.
```
