```python
sns.scatterplot(x='POL1', y='ACT104', data=df_diputados)

plt.title('Grafico de dispersion entre Rol con respecto al gobierno y Limitar a la oposici√≥n')
plt.xlabel('Rol con respecto al gobierno')
plt.ylabel('Limitar a la oposici√≥n')
plt.show()
```

```python
sns.regplot(x=df_diputados["POL1"], y=df_diputados["ACT104"], data=df_diputados)

plt.title('Regresi√≥n lineal simple entre Rol con respecto al gobierno y Limitar a la oposici√≥n')
plt.xlabel('Rol con respecto al gobierno')
plt.ylabel('Limitar a la oposici√≥n')
plt.show()
```

```python
x = df_diputados["POL1"]
y = df_diputados["ACT104"]

correlacion_pearson = np.corrcoef(x, y)[0, 1]

print("La correlaci√≥n de Pearson entre Rol con respecto al gobierno y Limitar a la oposici√≥n es:", correlacion_pearson)
```

```python
X = df_diputados[['POL1']]  
y = df_diputados['ACT104'] 
model = sm.OLS(y, X).fit()

r_squared = model.rsquared
print("El coeficiente de determinaci√≥n (R¬≤) es:", r_squared)
```

```python
covarianza = df_diputados[['POL1', 'ACT104']].cov()

print("La matriz de covarianza es:")
print(covarianza)
```
