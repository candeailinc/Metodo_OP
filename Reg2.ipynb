```python
#Hay gente que dice que sin partidos no puede haber democracia. ¬øHasta qu√© punto: mucho, algo, poco o nada, est√° Ud. de acuerdo con esta afirmaci√≥n?"
x = df_diputados["DEM6"]

print(x.head())
```

```python
#¬øqu√© grado de confianza, mucha, alguna, poca o ninguna, le merece su actuaci√≥n en la vida p√∫blica argentina? Los partidos pol√≠ticos
y = df_diputados["INST102"]

print(y.head())
```

```python
sns.scatterplot(x='DEM6', y='INST102', data=df_diputados)

plt.title('Grafico de dispersion entre Importancia de los partidos y Confianza en los partidos')
plt.xlabel('Importancia de los partidos')
plt.ylabel('Confianza en los partidos')
plt.show()
```

```python
correlacion_pearson = df_diputados["DEM6"].corr(df_diputados["INST102"])
print("La correlaci√≥n de Pearson entre Importancia de los partidos y Confianza en los partidos es:", correlacion_pearson)
```

```python
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

r_squared = model.rsquared
print("El coeficiente de determinaci√≥n (R¬≤) es:", r_squared)
```

```python
covarianza = df_diputados[['DEM6', 'INST102']].cov()

print("La matriz de covarianza es:")
print(covarianza)
```

```python
print(model.summary())
```

```python
plt.scatter(df_diputados["DEM6"], df_diputados["INST102"], label='Datos reales', color='blue')

intercepto, pendiente = model.params
x_vals = np.linspace(df_diputados["DEM6"].min(), df_diputados["DEM6"].max(), 100)
y_vals = intercepto + pendiente * x_vals

# Dibujar la recta de regresi√≥n en el gr√°fico de dispersi√≥n
plt.plot(x_vals, y_vals, color='red', label=f'Recta de regresi√≥n (pendiente={pendiente:.2f})')

# Personalizar el gr√°fico
plt.xlabel("Importancia de los partidos")
plt.ylabel("Confianza en los partidos")
plt.title('Regresi√≥n lineal')
plt.legend()

plt.show()
```

```python
print(f'Intercepto : {intercepto:.2f}')
print(f'Pendiente : {pendiente:.2f}')
```

INTERPRETACIONES: 
La correlaci√≥n entre las variables ‚Äúimportancia de los partidos‚Äù (DEM6) y ‚Äúconfianza en los partidos‚Äù (INST102) es negativa y muy baja (-0.0078), lo que indica una relaci√≥n inversa extremadamente d√©bil entre ambas variables.

El resultado de la regresi√≥n lineal (R¬≤) fue de 0.00006, lo que significa que el modelo no explica la variabilidad de la variable dependiente (confianza en los partidos) en funci√≥n de la variable independiente (importancia de los partidos). Adem√°s, el coeficiente de DEM6 es -0.0089, lo que significa que, por cada unidad de aumento en la importancia de los partidos, la confianza en los partidos disminuir√≠a m√≠nimamente (en 0.0089 unidades).

La matriz de covarianza muestra una covariaci√≥n baja y negativa entre DEM6 e INST102 (-0.005), lo que sugiere una leve asociaci√≥n inversa entre ambas variables. Ello coincide con el resultado de la correlaci√≥n de Pearson previamente explicada.

Respecto a la significatividad del modelo, de 0.935, indica que el modelo no es estad√≠sticamente significativo. Es decir, la variable DEM6 no tiene un impacto significativo en la variaci√≥n de INST102. 

El intervalo de confianza del modelo es del 95%, lo que indica que el efecto de la variable "importancia de los partidos" sobre "confianza en los partidos" podr√≠a ser nulo. Esto se refuerza por el hecho de que el margen de error incluye el valor 0, sugiriendo que no podemos afirmar con certeza que haya una relaci√≥n significativa entre ambas variables.

A modo de conclusi√≥n, se puede inferir que la relaci√≥n entre la "importancia de los partidos" y la "confianza en los partidos" es d√©bil y negativa. Adem√°s, la variable "importancia de los partidos" no resulta ser un factor explicativo relevante para la variable dependiente "confianza en los partidos pol√≠ticos".
