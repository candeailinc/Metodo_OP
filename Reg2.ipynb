```python
#Hay gente que dice que sin partidos no puede haber democracia. ¬øHasta qu√© punto: mucho, algo, poco o nada, est√° Ud. de acuerdo con esta afirmaci√≥n?"
x = df_diputados["DEM6"]

print(x.head())
```

```python
#¬øqu√© grado de confianza, mucha, alguna, poca o ninguna, le merece su actuaci√≥n en la vida p√∫blica argentina? Los partidos pol√≠ticos
y = df_diputados["INST102"]

print(y.head())
```

```python
sns.scatterplot(x='DEM6', y='INST102', data=df_diputados)

plt.title('Grafico de dispersion entre Importancia de los partidos y Confianza en los partidos')
plt.xlabel('Importancia de los partidos')
plt.ylabel('Confianza en los partidos')
plt.show()
```

```python
sns.regplot(x=df_diputados["DEM6"], y=df_diputados["INST102"], data=df_diputados)

plt.title('Regresi√≥n lineal simple entre Importancia de los partidos y Confianza en los partidos')
plt.xlabel('Importancia de los partidos')
plt.ylabel('Confianza en los partidos')
plt.show()
```

```python
correlacion_pearson = df_diputados["DEM6"].corr(df_diputados["INST102"])
print("La correlaci√≥n de Pearson entre Importancia de los partidos y Confianza en los partidos es:", correlacion_pearson)
```

```python
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

r_squared = model.rsquared
print("El coeficiente de determinaci√≥n (R¬≤) es:", r_squared)
```

```python
covarianza = df_diputados[['DEM6', 'INST102']].cov()

print("La matriz de covarianza es:")
print(covarianza)
```
